```python
import requests
from collections import defaultdict
import numpy as np
```

```python
UD_English_EWT = requests.get("https://raw.githubusercontent.com/UniversalDependencies/UD_English-EWT/master/en_ewt-ud-train.conllu")
lines = UD_English_EWT.text.split('\n')
lines[0:10]
```

```python
# Storing transition count
transition_counts = {}

# Going line by line
for i in range(len(lines)):
    line = lines[i]

    # Separating words by tab
    parts = line.split('\t')

    # Line index or number
    current_index_parts = parts[0].split('\'')

    # Correct lines are those which start with digits
    if not current_index_parts[0].isdigit():
        continue

    #Th eposition of pos tag
    current_pos = parts[3]
    current_index = int(current_index_parts[0])

    # Next line
    next_index = i + 1

    while next_index < len(lines):
        next_line = lines[next_index]
        next_parts = next_line.split('\t')
        if len(next_parts) >= 4 and '-' not in next_parts[0] and next_parts[0].isdigit():
            break
        next_index += 1

    if next_index < len(lines):
        next_pos = next_parts[3]
        transition_counts[(current_pos, next_pos)] = transition_counts.get((current_pos, next_pos), 0) + 1
```

```python
transition_probabilities = {transition: count / sum(transition_counts.values()) for transition, count in transition_counts.items()}
for transition, probability in transition_probabilities.items():
    print(f"{transition}: {probability}")
```
