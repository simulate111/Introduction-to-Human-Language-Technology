## Loading and reading a JSONL file

You can load a file from a given URL to the local drive using [`wget`](https://en.wikipedia.org/wiki/Wget), as shown below.

```python
!wget -nc https://raw.githubusercontent.com/simulate111/Introduction-to-Human-Language-Technology/main/topic2.jsonl
```

A file on the local drive in the simple [JSONL](https://jsonlines.org/) format can be read into a list of Python dictionary objects as follows:

```python
import json


data = []
with open('topic2.jsonl') as f:
  for l in f:
    d = json.loads(l)
    data.append(d)
```

Print a few of the first rows of data:

```python
for d in data[:5]:
  print(d)
```

## Plotting a word cloud

```python
!pip install --quiet wordcloud
```

Here's a small support function that uses the [`wordcloud`](https://pypi.org/project/wordcloud/) and [`matplotlib.pyplot`](https://matplotlib.org/3.5.3/api/_as_gen/matplotlib.pyplot.html) libraries to show a word cloud for a given text string. (You don't need to understand this is any detail, you can just use it as-is for visualization.)

```python
import matplotlib.pyplot as plt

from wordcloud import WordCloud


def show_word_cloud(text):
  cloud = WordCloud(width=1000, height=1000, background_color='white')
  cloud = cloud.generate(text)
  plt.figure(figsize=(10,10))
  plt.imshow(cloud)
  plt.axis('off')
```

```python
# Connect all sentences for visualization of the whole text together.
whole_text = ' '.join([d['text'] for d in data])
```

```python
# Word cloud of the whole text
show_word_cloud(whole_text)
```
